```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
d1 = -6
d2 = 10
d3 = 5
ds = [d1, d2, d3]
```

```python
c1 = '#f10700'
c2 = '#0000f1'
c3 = '#7cff79'
```

```python
plt.figure(figsize=(10,10))
plt.bar(x=[1,2,3], height=[d1, d2, d3], color=[c1, c2, c3])
plt.xticks([], [])
plt.yticks(fontsize=20)
plt.title('Estimated distances', fontsize=20)
plt.grid()
plt.savefig('distances.png')
plt.show()
```

```python
gamma = gamma = np.max(np.abs(ds))
exp_lst = [np.exp(-i/gamma) for i in ds]
probs = [exp/sum(exp_lst) for exp in exp_lst]
```

```python
probs
```

```python
plt.figure(figsize=(10,10))
plt.bar(x=[1,2,3], height=probs, color=[c1, c2, c3])
plt.xticks([], [])
plt.yticks(fontsize=20)
plt.title('Transformed probabilities', fontsize=20)
plt.grid()
plt.savefig('probabilities.png')
plt.show()
```

```python
np.cumsum(probs)
```

```python
plt.figure(figsize=(4,8))
plt.bar(x=[0,0,0], height=np.cumsum(probs)[::-1], color=[c1, c2, c3][::-1])
plt.xticks([], [])
plt.yticks(fontsize=20)
plt.title('Transformed probabilities', fontsize=20)
plt.grid()
plt.savefig('probabilitiesonebar.png')
plt.show()
```
